Удаление лишних символов

```python
#Удаление лишних симоволов
import re

### Define the regular expression pattern to match HTML tags
pattern1 = r"<.*?>"

# Define a function to remove HTML tags from a string
def remove_html_tags(text):
    return re.sub(pattern1, " ", text)

# Apply the function to each observation in the 'review' column
df_vtb['cleaned_review'] = df_vtb['review'].apply(lambda x: remove_html_tags(x))

### Define the regular expression pattern to match odd spaces
pattern2 = r" +"

# Define a function to remove odd spaces from a string
def remove_odd_spaces(text):
    return re.sub(pattern2, " ", text)

df_vtb['cleaned_review'] = df_vtb['cleaned_review'].apply(lambda x: remove_odd_spaces(x))

### Define the regular expression pattern
pattern3 = r"[^\w\s]"

# Define a function to remove all symbol except letters and digits
def remove_symbols(text):
    return re.sub(pattern3, " ", text)

df_vtb['cleaned_review'] = df_vtb['cleaned_review'].apply(lambda x: remove_symbols(x))
```

### 4. Найдите Топ-10 наиболее часто используемых слов в отзывах.

```python
from collections import Counter

# Сначала объединим все списки слов в один список всех слов
all_words = [word for review in df_vtb['cleaned_review'] for word in review]

# Затем используем Counter, чтобы подсчитать частоту встречаемости каждого слова
word_counts = Counter(all_words)

# Находим топ-10 наиболее часто встречающихся слов
top_10_words = word_counts.most_common(10)

# Выводим результат
print("Топ-10 наиболее часто используемых слов в отзывах:")
for word, count in top_10_words:
    print(f"{word}: {count} раз")
```

### 5. Разделите отзывы на положительные и отрицательные с использованием значения в поле «score» (например, 1,2- негативные, 4,5-позитивные).

```python
# Разделение отзывов на положительные и отрицательные
positive_reviews = df_vtb[df_vtb['score'].isin(['4', '5'])]
negative_reviews = df_vtb[df_vtb['score'].isin(['1', '2'])]
```

### 6. Найдите Топ-10 наиболее часто используемых прилагательных, используемых со словом банк для положительных/отрицательных отзывов.

```python
#вычленить пары, где перед словом банк идет слово
# Функция для определения части речи слова
def get_word_pos(word):
    return morph.parse(word)[0].tag.POS

# Функция для фильтрации пар по части речи первого слова
def filter_pairs(pair):
    first_word_pos = get_word_pos(pair[0])
    # Оставляем пары, где первое слово является прилагательным
    return first_word_pos == 'ADJF'

# Функция для создания пар слов 'какое-то слово' и 'банк' из списка токенов
def create_word_pairs(tokens):
    pairs = [(tokens[i], tokens[i+1]) for i in range(len(tokens)-1) if tokens[i+1] == 'банк']
    return pairs

# Создание нового столбца с парами слов 'какое-то слово' и 'банк'
positive_reviews['word_pairs'] = positive_reviews['cleaned_review'].apply(create_word_pairs)
negative_reviews['word_pairs'] = negative_reviews['cleaned_review'].apply(create_word_pairs)

# Удаление пар, где первое слово не является прилагательным
positive_reviews['word_pairs'] = positive_reviews['word_pairs'].apply(lambda pairs: [pair for pair in pairs if filter_pairs(pair)])
negative_reviews['word_pairs'] = negative_reviews['word_pairs'].apply(lambda pairs: [pair for pair in pairs if filter_pairs(pair)])
```

```python
# Положительные отзывы
# Сначала объединим все списки слов в один список всех слов
all_adjf_pos = [word for review in positive_reviews['word_pairs'] for word in review]

# Затем используем Counter, чтобы подсчитать частоту встречаемости каждого слова
adjf_counts_pos = Counter(all_adjf_pos)

# Находим топ-10 наиболее часто встречающихся слов
top_10_adjf_pos = adjf_counts_pos.most_common(10)

# Выводим результат
print("Топ-10 прилагательных, используемых со словом 'банк' в положительных отзывах:")
for adj, count in top_10_adjf_pos:
    print(f"{adj}: {count} раз")



# Отрицательные отзывы
# Сначала объединим все списки слов в один список всех слов
all_adjf_neg = [word for review in negative_reviews['word_pairs'] for word in review]

# Затем используем Counter, чтобы подсчитать частоту встречаемости каждого слова
adjf_counts_neg = Counter(all_adjf_neg)

# Находим топ-10 наиболее часто встречающихся слов
top_10_adjf_neg = adjf_counts_neg.most_common(10)

print("\nТоп-10 прилагательных, используемых со словом 'банк' в отрицательных отзывах:")
for adj, count in top_10_adjf_neg:
    print(f"{adj}: {count} раз")
```
